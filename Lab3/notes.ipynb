### 2a) Experimente o kill no lxsession. Nota: se não existir um lxsession encontre um processo com session no nome. (deve-se ver o que são os processos antes de os matar).


```bash
ps aux | grep lxsession
```

```bash
kill 8270
```

Acontece que sempre que verificamos o pid do processo "lxsession", este muda, que não permite que utilizamos o commando kill diretamente.

Assim, o que podemos tentar fazer é descubrir o pid e imediatamente enviar por um pipe para o kill.

```bash
kill $(ps aux | grep lxsession | awk '{print $2}')
```
